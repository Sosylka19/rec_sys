```python
import pandas as pd
import numpy as np
```

```python
data_vectors = pd.read_csv('./data/vectors.csv')
data_titles = pd.read_csv('./data/titles.csv')
```

```python
def func(arr: str) -> np.ndarray:
    c = arr
    b = c.split(', ')
    b[0] = b[0][1:]
    b[-1] = b[-1][:-1]
    b = np.array(list(map(float, b)))
    
    return b
```

```python
data_vectors['vectors'] = data_vectors['vectors'].map(lambda x: func(x))
```

```python
data_vectors['vectors'][0]
```

```python
vectors = np.stack(data_vectors['vectors'].values)
```

```python
import torch
```

```python
def recommend_movies(vectors, title, top_n=5):
  idx = data_titles[data_titles['original_title'] == title].index[0]
  # if not idx:
  #   return f"Film {title} not found"
  # idx = idx.index[0]
  target_vector = vectors[idx]

  similarities = np.dot(vectors, target_vector) / (np.linalg.norm(vectors, axis=1) * np.linalg.norm(target_vector))

  similar_indices = similarities.argsort()[-(top_n+1):-1][::-1]
  return similar_indices, similarities[similar_indices]
```

```python
similar_indices, scores = recommend_movies(vectors, 'Home Alone')
```
